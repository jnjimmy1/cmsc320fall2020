```python
import sqlite3
import pandas
from scipy import stats
import numpy as np

sqlite_file = 'lahman2014.sqlite'
conn = sqlite3.connect(sqlite_file)
```

# Part 1, Problem 1
My SQL creates a relation by multiplying the total_payroll of a team in a given year with their win ratio (as a percentage). Thus, this relation is about how much each team pays in total for their overall win ratio. 

In the Salaries table, there were some missing entities. For example, there was a teamID for each team, but no franchise ID. Meanwhile, in the Teams table, it included both the teamID and franchID's. As a result, I used the Inner Join to combine the two tables together. This way, only rows with both a matching year and teamID would be joined. Any row missing this data would be discarded.

Additionally, I removed the entries that had a yearID outside of the 1990-2014 range as discussed in Problem 2.

```python
relation_query = '''
SELECT Salaries.yearID, Salaries.teamID, sum(Salaries.salary) as total_payroll, 
Teams.W, Teams.L, Teams.W + Teams.L as num_games, ((W*1.0)/((W*1.0) + (L*1.0)) * 100) as win_ratio,
(sum(Salaries.salary)*1.0) * ((W*1.0)/((W*1.0) + (L*1.0))) as payroll_vs_win_ratio,
Teams.franchID 

FROM Salaries 

JOIN Teams ON (Salaries.yearID = Teams.yearID and Salaries.teamID = Teams.teamID)

GROUP BY Salaries.teamID, Salaries.yearID

ORDER BY Salaries.yearID
'''
#add back in WHERE Salaries.lgID == 'AL' ?
#pandas.read_sql()
rel_df = pandas.read_sql(relation_query, conn)
rel_df = rel_df[(rel_df['yearID'] >= 1990) & (rel_df['yearID'] <=2014)] #Limits the years from 1990-2014
rel_df.head()
```

# Part 2, Problem 2
The below code will create a plot detailing total payroll of each team from 1990 to 2014. Each line corresponds with a team within the database and their total payroll throughout time.

```python
import matplotlib.pyplot as plt

plt.title("MLB Team Payrolls Throughout Time")
plt.xlabel("Years")
plt.ylabel("Total Payrolls")
for team in rel_df['teamID'].unique():
    temp_df = rel_df[(rel_df['teamID'] == team)]
    plt.plot(temp_df['yearID'], temp_df['total_payroll'], label = team)
    plt.legend(loc = "upper left")
```

# Problem 3
The first plot details the average payroll of the entire league within a given year from 1985 to 2015. Clearly, this line increases over time which means the average team payroll increased over time. Thus, my statement about the "total payroll of each team increasing over time" is supported.

```python
# mean payroll over time
avg_payroll = []
for year in rel_df['yearID'].unique():
    summ = 0.0
    count = 0.0
    for index, rows in rel_df.iterrows():
        if rows['yearID'] == year:
            summ += rows['total_payroll']
            count += 1
    avg_payroll.append(summ / count)
#print(avg_payroll)
plt.title("Average Team Payroll Over Time")
plt.xlabel("Years")
plt.ylabel("Average Payroll")
plt.plot(rel_df['yearID'].unique(), avg_payroll)

    
```

# Problem 4
The below graph compares the average win ratio and average payroll within 4 time periods from 1984-2014. 
If some teams didn't exist during the time period, they were removed completely from the data for the graph.
First, I split the range of years into 5 intervals using pandas.cut. Then, for each interval, I went through and calculated the mean payroll as well as mean winratio for each team and plotted them. Finally, I made regression lines for each graph to help interpret the data.

```python
#split years into 5 intervals
years = pandas.cut(rel_df['yearID'].unique(), 5).unique()

#loop through each year
for chosen_year in years:
    plt.figure()
    x = []
    y = []
    
    #loop through each team
    for team in rel_df['teamID'].unique():
        #Get the avg payroll and winratios of each team within the time interval and plot them
        payroll_total = 0.0
        total_winratios = 0.0
        count = 0.0
        for index, rows in rel_df.iterrows():
            if (rows['yearID'] in chosen_year) & (rows['teamID'] == team):
                payroll_total += rows['total_payroll']
                total_winratios += rows['win_ratio']
                count += 1
        if count != 0: #Only plot if the team exists in the time period
            mean_payroll = payroll_total / count
            mean_winratio = total_winratios / count
            x.append(mean_payroll)
            y.append(mean_winratio)
            
            plt.annotate(team, xy=(mean_payroll, mean_winratio))
            plt.scatter(mean_payroll, mean_winratio, label = team)
    
    # Plot linear regression line of best fit   
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
    plt.plot(x, intercept + slope*np.asarray(x), 'r', label='fitted line')
    
    #plt.legend()
    plt.title("Win Ratio vs Payroll from " + str(chosen_year.left) + " to " + str(chosen_year.right))
    plt.xlabel("Average League Payroll")
    plt.ylabel("Average Win Ratios")
    #break
```

# Question 2
From the first graph, the slope of the regression line isn't as steep as the other graphs. Thus, in the time period 1989.976-1994.8, the payrolls didn't affect the win ratio as much. However, in subsequent years the higher the payroll was, the higher the winratio in general. Additionally, teams generally have similar payrolls too as there are several teams clumping vertically in the graphs.

Some teams frequently have high win ratios and vary in their payrolls. ATL consistently has one of the highest win ratios throughout all the time periods and their payroll has progressively been getting lower and lower. NYA is similar in that the team frequently has high win ratios, but their payroll is often the largest of all the teams too.

Oakland's efficiency has been high throughout time periods. In each graph, they have been above the regression line and frequently they've had one of the highest win rates with lower payrolls as seen in graphs 3 and 5. Their efficiency is definitely above average.

# Part 3, Problem 5
Using pandas, I quickly calculate the standard deviation of payroll for a given year and the mean for a given team within a given year. Then, I calculate the standardized payroll for a team in a specific year and add it to the dataframe.

```python
# Creates a new column named stnd_payroll and inputs the standardized payroll for a team in a given year
rel_df['stnd_payroll'] = 0.0
for year in rel_df['yearID'].unique():
    # Create a df based on year
    year_df = rel_df[rel_df['yearID'] == year]
    std_pay = year_df.std(axis=0).total_payroll
    mean_pay = year_df.mean(axis=0).total_payroll
    
    for team in rel_df['teamID'].unique():
        team_df = year_df[year_df['teamID'] == team]
        team_payroll = team_df.mean(axis=0).total_payroll
        standardized_payroll = ((team_payroll - mean_pay) * 1.0) / (std_pay * 1.0)
        
        # iterate through and set the standardized_payroll for the right row in rel_df
        for index, row in rel_df.iterrows():
            if (row['yearID'] == year) & (row['teamID'] == team):
                rel_df.at[index, 'stnd_payroll'] = standardized_payroll
                break
rel_df.head()
```

# Problem 6
The code is identical to Problem 4, but instead of using the regular payroll, I use the recently created standardized payroll.

```python
#split years into 5 intervals
years = pandas.cut(rel_df['yearID'].unique(), 5).unique()

#loop through each year
for chosen_year in years:
    plt.figure()
    x = []
    y = []
    
    #loop through each team
    for team in rel_df['teamID'].unique():
        #Get the avg stnd payroll and winratios of each team within the time interval and plot them
        stnd_payroll_total = 0.0
        total_winratios = 0.0
        count = 0.0
        for index, rows in rel_df.iterrows():
            if (rows['yearID'] in chosen_year) & (rows['teamID'] == team):
                stnd_payroll_total += rows['stnd_payroll']
                total_winratios += rows['win_ratio']
                count += 1
        if count != 0: #Only plot if the team exists in the time period
            mean_stnd_payroll = stnd_payroll_total / count
            mean_winratio = total_winratios / count
            x.append(mean_stnd_payroll)
            y.append(mean_winratio)
            
            plt.annotate(team, xy=(mean_stnd_payroll, mean_winratio))
            plt.scatter(mean_stnd_payroll, mean_winratio, label = team)
    
    # Plot linear regression line of best fit   
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
    plt.plot(x, intercept + slope*np.asarray(x), 'r', label='fitted line')
    
    #plt.legend()
    plt.title("Win Ratio vs Standardized Payroll from " + str(chosen_year.left) + " to " + str(chosen_year.right))
    plt.xlabel("Average League Standardized Payroll")
    plt.ylabel("Average Win Ratios")
    #break
```

# Problem 7
To make the plot, I loop through the entire df and add every entry's win ratio vs standardized payroll. That way, all the teams/year combination will be included. Then, I calculate a regression line based off this data.

```python
# For each team/year pair, plot the winratio vs payroll 
x = []
y = []
for index, row in rel_df.iterrows():
    x.append(row['stnd_payroll'])
    y.append(row['win_ratio'])
    plt.scatter(row['stnd_payroll'], row['win_ratio'])

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
plt.plot(x, intercept + slope*np.asarray(x), 'r', label='y={:.2f}x+{:.2f}'.format(slope,intercept))

plt.legend()
plt.xlabel('Standardized Average Payroll')
plt.ylabel('Average Win Ratio')
plt.title('Win Ratio vs Standardized Average Payroll')
```

# Problem 8
Using the efficiency equation given in the project description, I give each entry in the df an efficiency. Afterwards, I make a list of the suggested teams and plot their efficiencies throughout the years.

```python
rel_df['efficiency'] = 0.0
for index, row in rel_df.iterrows():
    rel_df.at[index, 'efficiency'] = row['win_ratio'] - (50+2.5*row['stnd_payroll'])

teams = ['OAK', 'BOS', 'NYA', 'ATL', 'TBA']
for team in teams:
    team_df = rel_df[rel_df['teamID'] == team]
    x = []
    y = []
    for year in rel_df['yearID'].unique():
        year_df = team_df[team_df['yearID'] == year]  
        for index, row in year_df.iterrows():
            x.append(year)
            y.append(row['efficiency'])
    plt.scatter(x, y, label = team)
        
plt.legend(loc="lower right")
plt.xlabel('Years')
plt.ylabel('Team Payroll Efficiency')
plt.title('Team Payroll Efficiency Over Time')
```
